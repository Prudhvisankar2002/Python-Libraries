```python
# pandas in NaN value?

import pandas as pd
import numpy as np
data=[10,22,43,np.nan,46,38]
df=pd.Series(data)
print(df)
```

```python
# create a series with nan string values?
import pandas as pd
import numpy as np
data=['prudhvi','sankar','ramana']
df=pd.Series(data)
print(df)
```

```python
#pandas in series on values attributes?
import pandas as pd
result=[22,55,43,76,44,99,45,66]
df=pd.Series(result)
print(df)
print(df.values)
```

```python
#pandas in series on index attributes?
import pandas as pd
result=[22,55,43,76,44,99,45,66]
df=pd.Series(result)
print(df)
print(df.index)
```

```python
#pandas in series on index on string attributes?
import pandas as pd
result=[22,55,43,76,44,99,45,66]
i=['a','b','c','d','e','f','g','h']
df=pd.Series(result,index=i)
print(df)
print(df.index)
```

```python
#pandas in series on dtype attributes?
import pandas as pd
result=[22,55,43,76,44,99,45,66]
series=pd.Series(result)
print(series)
print(series.dtype)
```

```python
#pandas in series on dtype attributes?
import pandas as pd
result=[22,55,'rakesh',76,'ramaya',99,45,66]
series=pd.Series(result)
print(series)
print(series.dtype)
```

```python
#pandas in series on size attributes?
import pandas as pd
result=[22,55,47,76,99,45,66]
series=pd.Series(result)
print(series)
print(series.size)
```
