```python
# Write a program to create line gragh witht using the matplotlib-labrarie?

import matplotlib.pyplot as plt
x=[1,4,9]
y=[2,5,8]
plt.plot(x,y,marker="*")
plt.title("line graph representation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# write a simple line program chart?

import matplotlib.pyplot as plt
x=[1,6]
y=[2,7]
plt.plot(x,y,marker="*")
plt.show()
```

```python
# write a double lines program?

import matplotlib.pyplot as plt
product=[1,6,7,4,8]
sales=[1,5,6,8,6]
buy=[4,5,7,8,9]
plt.plot(product,sales,marker="*")
plt.plot(product,buy,marker="*")
plt.title("Two lines representation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# write a program for create a BAr_chart?

import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may","jun","jul","aug","sept","oct","nov","dec"]
sales=[12,10,30,34,44,67,45,67,93,56,22,78]
plt.bar(months,sales)
plt.title("bar chart representation")
plt.xlabel("months")
plt.ylabel("sales")
plt.show()
```

```python
# write a program for create a BAr_chart in horizental?

import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may","jun","jul","aug","sept","oct","nov","dec"]
sales=[12,10,30,34,44,67,45,67,93,56,22,78]
plt.barh(months,sales)
plt.title("bar chart representation")
plt.xlabel("months")
plt.ylabel("sales")
plt.show()
```

```python
# write a program for create a Bar_chart using width ?

# width values are presentation to asign with :-decimal values(1.1, 0.1, 1.5, 0.5, 0.7)
import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may","jun","jul","aug","sept","oct","nov","dec"]
sales=[12,10,30,34,44,67,45,67,93,56,22,78]
plt.bar(months,sales,width=0.5)
plt.title("bar chart representation")
plt.xlabel("months")
plt.ylabel("sales")
plt.show()
```

```python
# write a program for create a Histogram using  with (bins)?
# bins are used with integer values like:10,20,30,40,50

import matplotlib.pyplot as plt
sales=[12,10,30,34,44,67,45,67,93,56,22,78]
plt.hist(sales,bins=20)
plt.title("Histogram representation")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# write a program to  create a pie chart?

import matplotlib.pyplot as plt
months=["jan","feb","mar","apr","may","jun","jul","aug","sept","oct","nov","dec"]
sales=[12,10,30,34,44,67,45,67,93,56,22,78]
plt.pie(sales,labels=months)
plt.title("pie chart representation")
plt.axis('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
developers=['prudhvi','sankar','ram','ravi']
salary=[100000,20000,30000,40000]
plt.pie(salary,labels=developers,shadow=True,explode=(0.1,0.1,0.1,0.1),autopct='%1.1f%%')
plt.title("pie chart representation")
plt.axis('equal')
plt.legend()
plt.show()
```

```python
# write a program create a scatter plot chart?

import matplotlib.pyplot as plt
area=[10,20,30,40,50]
rice_packets=[25,50,75,100,125]
plt.scatter(area,rice_packets,marker="*",c="blue")
plt.title("scatter plot representation")
plt.xlabel("area")
plt.ylabel("rice_packets")
plt.show()
```

```python
# write a program to createa BOx plot?

import matplotlib.pyplot as plt
rice_packets=[25,50,75,100,125]
plt.boxplot(rice_packets)
plt.title("BOX Plot representation")
plt.xlabel("x-axis")
plt.ylabel("x-axis")
plt.show()
```

```python
# write a program to create a BOx plot with out of range ?

import matplotlib.pyplot as plt
rice_packets=[-200,-50,25,50,75,100,125,400]
plt.boxplot(rice_packets)
plt.title("BOX Plot representation")
plt.xlabel("x-axis")
plt.ylabel("x-axis")
plt.show()
```
