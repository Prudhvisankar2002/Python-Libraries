```python
  # numpy on methods

  #//methods are predefined methods

  #They are:-sum(),min(),max(),sort(),reshape(),flatend(),trace()
```

```python
import numpy as np                                         #SUM()-->methode
data=[[12,23,45],[11,56,90],[55,67,45]]
find=np.array(data)
print(find)
print()
print('Total sum of the data is:',find.sum())
```

```python
import numpy as np                                         #MIN()-->methode
data=[[12,23,45],[11,56,90],[55,67,45]]
find=np.array(data)
print(find)
print()
print('minimum value of the data is:',find.min())
```

```python
import numpy as np                                         #max()-->methode
data=[[12,23,45],[11,56,90],[55,67,45]]
find=np.array(data)
print(find)
print()
print('maximum value of the data is:',find.max())
```

```python
import numpy as np                                         #RESHAPE()-->methode
data=[[12,23,45],[11,56,90],[55,67,45],[38,25,78]]                    #reshape() used enter with parameter inside the paranthasis()
find=np.array(data)
print(find)
print()
print('Reshape of the data is:',find.reshape(6,2))
```

```python
import numpy as np                                         #SORT()-->methode
data=[[55,23,5],[11,10,7],[55,67,45]]
find=np.array(data)
print(find)
print()
a = find.sort()
print(a)
```

```python
import numpy as np                                         #FLATTEN()-->methode
data=[[12,23,45],[11,56,90],[55,67,45]]
find=np.array(data)
print(find)
print()
print('data change into one dimension:',find.flatten())
```

```python
import numpy as np                                         #trace()-->methode
data=[[12,23,45],[11,56,90],[55,67,45]]
find=np.array(data)
print(find)
print()
print('diagonals of the data is:',find.diagonal())
print()
print('diagonal sum is:',find.trace())
```
