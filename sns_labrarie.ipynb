```python
# write a program to print the unique values for sex (gender) cloumn?

import seaborn as sns
df=sns.load_dataset('titanic')
print(df['sex'].unique())
```

```python
# write a program to print the unique values for surviveded  and un_servived counts?

import seaborn as sns
df=sns.load_dataset('titanic')
print(df['survived'].value_counts())
```

```python
# write a program to print bar_plot for sex and age cloumns?

import seaborn as sns
import matplotlib.pyplot as plt
df=sns.load_dataset('titanic')
sns.barplot(x='sex',y='age',data=df)
plt.show()
```
