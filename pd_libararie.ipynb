```python
# Write a program create a empty data frame?

import pandas as pd
df=pd.DataFrame()
print(df)
```

```python
# Write a program create a dataframe by using single list?

import pandas as pd
data=[55,67,45,68,89,23]
df=pd.DataFrame(data)
print(df)
```

```python
# Write a program create a dataframe by using single list of list?

import pandas as pd
#data creation list in list
data=[ [101,'prudhvi',10000.0],
       [102,'sankar',20000.0],
       [103,'ramana',30000.0]
     ]
df=pd.DataFrame(data,columns=['emp_id','emp_name','emp_salary'])
print(df)
```

```python
# Write a program create a dataframe by using DICTIONARY?

import pandas as pd
data={
      'names':['raja','rani','bantu'],
      'Age':[33,44,55],
      'city':['hyderabad','delhi','chennai']
     }
df=pd.DataFrame(data)
print(df)


```

```python
# write a program for empty Series object?

# Series is pre_defined object in pandas.

import pandas as pd
df=pd.Series()
print(df)

```

```python
# write a program for using list Series object and assign a name?

import pandas as pd
data=[10,20,30,40,50]
df=pd.Series(data,name='marks')
print(df)
```

```python
# write a program for using list Series object and assign a name?
import pandas as pd
data=['raja','rani','bantu','krishna']
df=pd.Series(data,name='students')
print(df)
```

```python
# writa a program to print creating the Series from array?

import pandas as pd
import numpy as np
data=(23,55,86,35,25,56)
data1=np.array(data)
s=pd.Series(data1)
print(s)
```

```python
# writa a program to print create to using the Series from array?

import pandas as pd
import numpy as np
data=['ramya','rakesh','ravi','sreja','sankar']
values=np.array(data)
string=pd.Series(values,name='emp_names')
print(string)
```
