```python
def show(a,b,c,d):                #1
    tot=a+b
    return tot
tot=show(1,2,3,4)
print(tot)
```

```python
def display():                    #2
    p=5
    q=7
    return p,q
a,b=display()
print(a,b)
```

```python
def emical(b,t,r):                #3
    intrest=(b*t*r)/100
    return intrest
intamt=emical(100,2,2)
print(intamt)


```

```python
def payment(cardno,name):          #4
    print(cardno)
    print(name)
    return True
res=payment(111,"sankar")
print(res)
```

```python
def caltotal(a,b,c):               #5
    total=a+b+c
    return total
def calavg(t):
    avg=t/3
    return avg
tot=caltotal(1,2,3,)
average=calavg(30)
print(tot)
print(average)


```
